# 20. QLS - Hypothesis Testing

Statistical inference, making predictions about a large group based on smaller samples, is broken down into two segments: **estimation**, and **hypothesis testing**. Estimation provides values for metrics such as mean and variance, with a given confidence interval. A confidence interval provides a region in which the true value of the estimated parameter can be found. More information can be found on [this](./quantopian_lectures/21QLS_Confidence-Intervals.ipynb) lecture.

For example, the estimate mean of a sample is 100, with a confidence interval of 90%; meaning 110 at a 95% confidence interval. This isn't to say that the true population mean is within 90 and 110 with a 95% probability, as the true mean is a fixed value, and the probability is 100% or 0%. Instead, what this means is that over many computations at a 95% confidence interval (assuming underlying assumptions about distributions hold) the population mean will be in the interval 95% of the time.

This portrays an idea of the specific characteristics that an overall population may exhibit, given a sample. Hypothesis testing provides a different focus, detailing a framework for statistical testing of hypothesized values. By making an assertion of what a value should be, a testable hypothesis is created.

Statistical tests are designed such that if all the pre-requisite conditions are met, the right answer about the data should be achieved a certain percentage of time. When a hypothesis is accepted as true based on a test, that does not mean it's definitely true. It just means the probability of being **wrong** is known.

## The Null and Alternative Hypothesis

The first thing to be introduced is the null hypothesis, written as $H_0$. The null hypothesis is the default case, generally reflecting the common conception. The alternative hypothesis is the one to be tested.

### Examples

The alternative hypothesis, $H_A$, is that more than 10 pairs of shoes are owned. The null hypothesis, $H_0$, is that more than 10 pairs of shoes are **not** owned.
The alternative hypothesis, $H_A$, is that eating pizza is related to obesity. The null hypothesis, $H_0$, is that it's not.
The alternative hypothesis, $H_A$, is that Microsoft's mean returns are greater than 0. The null hypothesis, $H_0$, is that they are less than or equal to 0.

### Difficulty of Testing

Some hypotheses are easier to test that others. For instance, the alternative hypothesis, "more than 10 pairs of shoes are owned," and the accompanying null hypothesis that they are not, is easily tested by counting pairs. However, a perfect answer might not be achieved all the time in this case, since measurement error might be present in the counting.

On the other hand, the hypothesis, "the number of insect species is greater than the number of stars in the universe," would be more difficult to test and requires lots of data gathering.

### Hypotheses Must be Testable

A hypothesis can't be vague, otherwise how is it tested. E.g. "momentum trading is a good way to make money," is not really testable. What does 'good' mean? What type of momentum trading is being discussed? Hypotheses should be specific and the type of test needed should be known.

## How to Perform Hypothesis Testing

These are the main steps in hypothesis testing:
1. State the hypothesis and the alternative to the hypothesis.
2. Identify the appropriate test statistic and its distribution; ensure that any assumptions about the data are met (stationarity, normality, etc.)
3. Specify the significance level, $\alpha$.
4. From $\alpha$ and the distribution, compute the 'critical value'.
5. Collect the data and calculate the test statistic.
6. Compare test statistic with critical value and decide whether to accept or reject the hypothesis.

First, state the hypothesis to be tested. This is done by identifying a **null hypothesis**, and an **alternative hypothesis**. The null hypothesis, $H_0$, is the one to test, while the alternative hypothesis, 
$H_A$, is the accepted hypothesis in the case where $H_0$ is rejected.

To test whether the mean return of Microsoft's stock is positive: the parameter that is being tested is denoted by $\theta$, and the proposed value of the parameter is denoted by $\theta_0$, which in this case is equal to $0$. So $H_0 = \theta$, that the returns are negative, and $H_A$ is $\theta \neq \theta_0$. Including this formation, there are three possible ways to formulate the null and alternative hypotheses:

1. $H_0: \theta = \theta_0$ versus $H_A: \theta \ne \theta_0$ (A "not equal to" alternative hypothesis).
2. $H_0: \theta \leq \theta_0$ versus $H_A: \theta > \theta_0$ (A "greater than" alternative hypothesis).
3. $H_0: \theta \geq \theta_0$ versus $H_A: \theta < \theta_0$ (A "less than" alternative hypothesis).

In this case, where the returns of MSFT are being tested, $\theta = \mu_{\text{MSFT}}$, representing the stock's mean returns. Since what's being tested is whether the returns are positive or negative, $\theta_0 = 0$. The example follows the first formulation of a hypothesis test. This is a **two-tailed hypothesis test**. The second and third formulations are examples of a **one-tailed hypothesis test**. With a one-tailed test, the null is rejected in favour of the alternate ONLY if the data indicates that $\theta_0 < \theta_n < \theta_0$. Meanwhile, a two-tailed test rejects the null in favour of the alternate if the data indicates that $\theta$ is EITHER greater than, or less than $\theta_0$.

The hypothesis, in qualitative terms, would look like:

\begin{align}
    H_{0} &:& \text{The mean return on MSFT stock is $0$}\\
    H_{A} &:& \text{The mean return on MSFT stock is not $0$}
\end{align}

The null and alternative hypothesis must be complementary to each other. Between them, they must cover all the values of $\theta$.

The most common type of hypothesis test is the two-tailed, because it presents a neutral view. The one-tailed hypothesis tests are less neutral as they reflect the thoughts of the tester. Intuitively, one-tailed tests are often used to test for 'hoped for' results, or results that the tester has an idea about.

```python
import numpy as np
import pandas as pd
import seaborn as sb
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.set_style("whitegrid", {"grid.color":".6", "grid.linestyle":":"})
import yfinance as yf
```

```python
start = "2015-01-01"
end = "2016-01-01"

pricing_sample = yf.download("MSFT", start, end)['Adj Close']
returns_sample = pricing_sample.pct_change()[1:]

sb.lineplot(pricing_sample).set(xlabel="Date", ylabel="Price");
```

Here are the returns.

```python
sb.lineplot(returns_sample).set(xlabel="Date", ylabel="Returns");
```

## Why is This All Necessary?

Why not just take the mean returns of Microsoft and check if they are $>0$? Because the returns can only be sampled on a limited time period, not to mention that the underlying data generating process is unknown. Since only a sample is being observed, that sample may or may not reflect the true state of the underlying process. Because of that, statistical tests need to be used.

So, an appropriate **test statistic** and its corresponding probability distribution need to be identified. A test statistic usually takes the following form:

$$\text{Test Statistic } = \frac{\text{Sample statistic } - \text{ Value of the population parameter under }H_{0} (\theta_{0})}{\text{Standard error of the sample statistic}}$$

It's calculated based on sample data, and it's compared to its probability distribution to determine whether to reject or not the null hypothesis. Since the mean return of MSFT is being tested, the sample statistic is the sample mean $\overline{x}$. If the population standard deviation $\sigma_{x}$ is known, the standard error is $\frac{\sigma_{x}}{\sqrt{n}}$, Otherwise it's calculated as $\frac{S_{x}}{\sqrt{n}}$, where $S_{x}$ is the sample quasi-standard deviation. So, with that, the test statistic is calculated as:

$$T = \frac{\overline{x} - \mu_{0}}{\frac{S_{x}}{\sqrt{n}}}$$

The four most common distributions for test statistics are:
- The $t$-distribution ($t$-test).
- The standard/normal distribution ($z$-test).
- The chi-square ($\chi^{2}$) distribution ($\chi^{2}$-test).
- The $F$-distribution ($F$-test).

After the appropriate test statistic and probability distribution are identified, the tester must specify the **significance level** ($\alpha$) of the test. The values for which the test statistic is compared to, in order to reject or not reject $H_{0}$ are determined based on $\alpha$.

|                    |   **$H_{0}$ is true**   |         **$H_{0}$ is false**         |
|:------------------:|:-----------------------:|:------------------------------------:|
| **Accept $H_{0}$** |  No error ($1-\alpha$)  |        Type II error ($\beta$)       |
| **Reject $H_{0}$** | Type I error ($\alpha$) | No error ($1-\beta = \text{ power}$) |

The significance level corresponds to the probability of a Type I error (a 'false positive') happening. The probability of a Type II error (a 'false negative') is denoted by $\beta$. By decreasing the probability of a Type I error, the probability of a Type II error is increased, resulting in a tradeoff. The only way to reduce both at the same time is to increase the sample size.

The conventional significance levels are $0.1$, $0.05$, and $0.01$. Rejecting the null at each level means there is some evidence that it's false, strong evidence, and very strong evidence respectively.

### Critical Value

Now the critical value, or rejection point, must be calculated. The **critical value** for a test statistic is the value that is compared against the test statistic to decide whether to reject the null hypothesis. If the null is rejected, then the result is <u>statistically significant</u>, while if the null fails to be rejected, the result is <u>not statistically significant</u>.

The critical value of a test is determined based on the $\alpha$ of the hypothesis test as well as the corresponding distribution. Given $\alpha = 0.05$, and a one-sided $z$-test, there are two different ways to interpret the critical value:

- If the test is right-tailed ($H_{0}: \theta \le \theta_{0}, H_{A}: \theta > \theta_{0}$) at $\alpha = 0.05$, the critical value is $z_{0.05} = 1.645$. So, the null hypothesis is rejected if $T > 1.645$.
- If the test is left-tailed ($H_{0}: \theta \ge \theta_{0}, H_{A}: \theta < \theta_{0}$) at $\alpha = 0.05$, the critical value is $z_{0.05} = -1.645$. So, the null hypothesis is rejected if $T < -1.645$.

For a two-sided test it's a slightly different situation. Naturally, there are two rejection regions, negative and positive. Given $\alpha = 0.05$, and the fact that the total probability of a Type I error must sum up to $0.05$ in this scenario, the rejection points must split $\alpha$. So, the rejection points for a $z$-test are $z_{\frac{0.05}{2}} = \pm 1.96$ respectively. Thus, the null is rejected if $-1.96 < T < 1.96$ doesn't hold.

When carrying out a hypothesis test, another method to determine the result is using $p$-values. A **$p$-value** is the minimum level of significance where the null hypothesis can be rejected. Often, a misleading interpretation of $p$-values is that they are "the probability that the null hypothesis is false". But, a $p$-value only makes sense when being compared to the significance value. If a $p$-value is less than $\alpha$, the null is rejected, otherwise it isn't. Lower $p$-values don't make something more "statistically significant".

Significance values tie nicely into confidence intervals, which are covered more in-depth in the confidence intervals lecture. A confidence interval provides an estimate for a parameter's possible range in values given a certain significance level. For example, if a 99% confidence interval for the mean returns of MSFT was $(-0.0020, 0.0023)$, that would mean that there's a 99% chance that the true value of the mean is within that interval.

```python
from scipy.stats import norm

# generate a range of x values
x = np.linspace(-3, 3, 100)

# calculate the probability density function (PDF) for the standard normal distribution
pdf = norm.pdf(x)

# create the plot
ax = sb.lineplot(x=x, y=pdf, label='Normal Distribution', color='blue')

# calculate the critical values for a 95% confidence level (alpha = 0.05)
alpha = 0.05
critical_value = norm.ppf(1 - alpha / 2)

ax.fill_between(x, pdf, where=(x < -critical_value) | (x > critical_value), color='blue', alpha=0.5, label='Rejection Region (α = 0.05)')
ax.set_title('Rejection regions for a two-tailed hypothesis test at 95% confidence')
ax.set_xlabel('Z-score')
ax.set_ylabel('Probability Density')
ax.legend()

sb.despine()
```

Now, the relevant data for the test must be collected to calculate the test statistic for a two sided, 5% significance test. Note how any negative characteristics of the data would render the hypothesis test invalid. For the case of MSFT returns, issues of time-period bias, or look-ahead bias could arise. As always with historical data, it might result in a specific test result that may not hold in the future.

Here, the test statistic is calculated:

```python
n = len(returns_sample['MSFT'])
test_statistic = ((returns_sample['MSFT'].mean() - 0)/(returns_sample['MSFT'].std()/np.sqrt(n)))

print("Test Statistic (T): ", test_statistic)
```

In order to take the statistical decision for the test, the test statistic must be compared to the critical value. The test statistic, as outlined above, is between the two critical values for a 95% two-tailed $z$-test. So, in this example, the null hypothesis is not rejected.

To determine the result of this hypothesis test using $p$-values, it would be calculated in the following way:

```python
from scipy.stats import t
```

```python
p_value = 2 * (1- t.cdf(test_statistic, n-1))
print("P-value: ", p_value)
```

Since the $p$-value is greater than the significance level, $\alpha = 0.05$, the null hypothesis is not rejected.

After making the statistical decision, it must be translated into real life. Often, the conclusion might not be something actionable, but the results can hyave other implications. In the case of the example outlined previously, the daily returns of MSFT in 2015 are not significantly different from $0$.

## Hypothesis Testing on Means

A $z$-distribution, or a standard normal distribution, is an essential probability distribution in finance. It's good when things are normally distributed since that entails a lot of useful properties. On top of that, many fundamental methods require an assumption of normality. However, in most cases a $z$-distribution will be inappropriate for the data. This is because it's rare to have the true population values (mean and variance) of the data, so approximations need to be made. In those cases, the $t$-distribution is a better fit, as it's more forgiving when it comes to small sample sizes. It has fatter tails and a lower peak, which provides more flexibility when compared to a $z$-distribution.

Both the $t$ and $z$-distributions rely on an underlying assumption of normality, which is usually the case when analyzing financial data. Therefore, in addition to testing individual data points, it might make sense to compare between two or more values. A hypothesis test can be used to determine whether the means of <u>several data-sets</u> are statistically different from one another.

The following uses a $t$-distribution to compare the mean returns of the S&P500, and AAPL with a hypothesis test to see if the differences are statistically significant or not.

```python
start = "2015-01-01"
end = "2016-01-01"

pricing_sample2 = yf.download(["SPY", "AAPL"], start, end)['Adj Close']
returns_sample2 = pricing_sample2.pct_change()[1:]

sb.lineplot(returns_sample2).set(xlabel="Date", ylabel="Returns");
```

While the returns <u>appear</u> to have the same mean, there is still not enough evidence to conclude that with certainty. So a hypothesis test is used to ground suspicions in a statistical basis.

When comparing two means, the hypothesis tests are stated as the following:
- Two-tailed: $H_{0}: \mu_{1} - \mu_{2} = \theta_{0}, H_{A}: \mu_{1} - \mu_{2} \ne 0$.
- Right-tailed: $H_{0}: \mu_{1} - \mu_{2} \le \theta_{0}, H_{A}: \mu_{1} - \mu_{2} > 0$.
- Left-tailed $H_{0}: \mu_{1} - \mu_{2} \ge \theta_{0}, H_{A}: \mu_{1} - \mu_{2} < 0$.

Where $\mu_{1}, \mu_{2}$ are the respective sample means of SPY and AAPL, and $\theta_{0}$ is the parameter of the test. A two-tailed hypothesis test will be used to verify the equality of the two returns. Assuming that the population variances are equal, the test statistic is calculated as:

$$T = \frac{(\overline{x}_{1} - \overline{x}_{2}) - (\mu_{1} - \mu_{2})}{\sqrt{S_{p}^{2} \cdot \frac{1}{n_{1}} + \frac{1}{n_{2}}}}$$

With $S_{p}^{2} = \frac{(n_{1} - 1)S_{x_{1}}^{2} - (n_{2} - 1)S_{x_{2}}^{2}}{n_{1} + n_{2} - 2}$ as the estimator of the common variance, known as the **pooled variance**, and $n_{1} + n_{2} - 2$ as the number of degrees of freedom ($n_{1} - 1$ and $n_{2} - 1$ for each dataset). A typical $t$-test on a mean assumes that all variances are equal with underlying normal distributions. If the assumption is that the variances are <u>not equal</u>, then the test statistic should be:

$$T = \frac{(\overline{x}_{1} - \overline{x}_{2}) - (\mu_{1} - \mu_{2})}{\sqrt{\frac{S_{1}^{2}}{n_{1}} + \frac{S_{2}^{2}}{n_{2}}}}$$

Where the number of degrees of freedom used to find the critical statistic is calculated as,

$$df = \frac{\left( \frac{S_{1}^{2}}{n_{1}} + \frac{S_{2}^{2}}{n_{2}} \right)^{2}}{\frac{\left( \frac{S_{1}^{2}}{n_{1}} \right)^{2}}{n_{1} - 1} + \frac{\left( \frac{S_{2}^{2}}{n_{2}} \right)^{2}}{n_{2} - 1}}$$.

This preserves the underlying normality of the data being tested while accounting for differing variances. Calculating the statistic this way removes a lot of problems that occur if there are unequal variances, especially if the sample sizes also differ. This specific case of a $t$-test is referred to as "[Welch's unequal variances $t$-test](https://en.wikipedia.org/wiki/Welch%27s_t-test)".

For this example, the assumption is that the variances of SPY and AAPL returns are different (i.e. AAPL will be riskier than SPY), so the second test statistic should be used.

```python
mu_spy = returns_sample2["SPY"].mean()
mu_aapl = returns_sample2["AAPL"].mean()
std_spy = returns_sample2["SPY"].std()
std_aapl = returns_sample2["AAPL"].std()
n_spy = len(returns_sample2["SPY"])
n_aapl = len(returns_sample2["AAPL"])

test_statistic2 = ((mu_spy - mu_aapl) - 0) / np.sqrt((std_spy**2 / n_spy) + (std_aapl**2 / n_aapl))
df = ((std_spy**2 / n_spy) + (std_aapl**2 / n_aapl))**2 / (
    ((std_spy**2 / n_spy)**2 / (n_spy - 1)) + ((std_aapl**2 / n_aapl)**2 / (n_aapl - 1))
)


print("Test Statistic (T): ", test_statistic2)
print("Degrees of freedom: ", df)
```

By looking at a [$t$-table](https://en.wikipedia.org/wiki/Student%27s_t-distribution#Table_of_selected_values), the critical values for a two-sided hypothesis test are $-1.96, 1.96$ respectively. Since the test statistic is between those values, the null hypothesis is not reject, meaning the difference between SPY and AAPL returns is not significantly different from $0$.

## Hypothesis Testing on Variance

To test the variance of a population, a the $z$ or $t$-distribution can't be used. Variances, by definition, must be greater than or equal to $0$. The latter distributions allow for negative values, making them unviable for this type of hypothesis test. Risk is quantified in terms of variance, so this method of hypothesis testing is a handy addition to a financial toolbox.

The most common distributions are the $\chi^{2}$ distribution for single variance tests, and the $F$ distribution for comparisons of variance, since they are lower bound by $0$. 

A single variance test can take on three forms:
- $H_{0}: \sigma^{2} = \sigma_{0}^{2}, H_{A}: \sigma^{2} \ne \sigma_{0}^{2}$.
- $H_{0}: \sigma^{2} \le \sigma_{0}^{2}, H_{A}: \sigma^{2} > \sigma_{0}^{2}$.
- $H_{0}: \sigma^{2} \ge \sigma_{0}^{2}, H_{A}: \sigma^{2} < \sigma_{0}^{2}$.

The $\chi^{2}$ distribution is a <u>family</u> of functions with each different formulation determined by the number of degrees of freedom. That is to say, the shape of the distribution will be different for different $n$ degrees of freedom.

```python
from scipy.stats import chi2
x = np.linspace(0, 8, 1000)
y_1 = chi2.pdf(x, 1)
y_2 = chi2.pdf(x, 2)
y_3 = chi2.pdf(x, 3)
y_4 = chi2.pdf(x, 4)
y_6 = chi2.pdf(x, 6)
y_9 = chi2.pdf(x, 9)

sb.lineplot(x=x, y=y_1, label="n = 1")
sb.lineplot(x=x, y=y_2, label="n = 2")
sb.lineplot(x=x, y=y_3, label="n = 3")
sb.lineplot(x=x, y=y_4, label="n = 4")
sb.lineplot(x=x, y=y_6, label="n = 6")
sb.lineplot(x=x, y=y_9, label="n = 9")
sb.lineplot().set(xlabel="Chi-squared", ylabel="Probability Density", title="Chi-squared Distribution with n degrees of freedom");
```

The $\chi^{2}$ test statistic is calculated as:

$$\chi^{2} = \frac{(n-1) S^{2}}{\sigma_{0}^{2}}$$

Where $S^{2}$ is the sample variance and $n$ is the size of the dataset. The number of degrees of freedom is $n - 1$, and this is used along with the test statistic to determine the critical value(s) of the $\chi^{2}$ hypothesis test.

```python
sb.lineplot(returns_sample).set(xlabel="Date", ylabel="Price");
```

Now, a $\chi^{2}$ test can be used for the value of the variance of MSFT stock. Assume an $\alpha = 0.01$, and a test on whether the variance is less than or equal to $0.0001$ (that is to say, that the risk is less than or equal to $0.01$).

$$H_{0}: \sigma^{2} \le 0.0001, H_{A}: \sigma^{2} > 0.0001$$

So, to calculate the test statistic:

```python
test_statistic3 = ((len(returns_sample['MSFT']) - 1) * returns_sample['MSFT'].std()**2)/(0.0001)
print("Test statistic (T): ", test_statistic3)
```

```python
critical_value3 = chi2.ppf(0.99, len(returns_sample['MSFT']) - 1)
print("Critical value, given a = 0.01, and n = 252: ", critical_value3)
```

Since $792.0998 > 304.9395$, the null hypothesis is rejected in favor of the alternative, and $\sigma^{2} > 0.0001$.

### Comparing Two Variances

To compare the variances of two <u>different</u> things, the $F$ distribution must be used.

The $F$ distribution is similar to the $\chi^{2}$ in that it's asymmetrical and bounded below by $0$. The $F$ distribution is defined with two different values of degrees of freedom. When it comes to hypothesis testing, each one correlates to <u>one</u> of the factors being compared. 

An $F$ distribution can be constructed from two separate $\chi^{2}$ distributions. $X$ is a $F$ random variable if it can be written as $X = \frac{\frac{Y_{1}}{n_{1}}}{\frac{Y_{2}}{n_{2}}}$, where $Y_{1}$ and $Y_{2}$ are $\chi^{2}$ random variables with degrees of freedom $n_{1}$ and $n_{2}$ respectively.

An $F$ random variable is essentially a ratio of variances. So, constructing the $F$ test statistic is done by taking the ratio of the sample variances of the data that is being tested.

$$F = \frac{\sigma_{1}^{2}}{\sigma_{2}^{2}}$$

Comparing AAPL and SPY to check whether their variances are the same (a two sided hypothesis test) could be a useful exercise. Assume $\alpha = 0.05$. First, the data must be gathered to calculate the test statistic:

```python
std_spy = returns_sample2["SPY"].std()
std_aapl = returns_sample2["AAPL"].std()

print("SPY's standard deviation is: ", std_spy)
print("AAPL's standard deviation is: ", std_aapl)
```

Note that the standard deviation of AAPL is greater than the one of SPY. Hence, $\sigma_{1}^{2}$ corresponds to the variance of AAPL, and $\sigma_{2}^{2}$ to SPY.

```python
test_statistic4 = (std_aapl/std_spy)**2
print("Test statistic (T): ", test_statistic4)
```

```python
n_aapl_1 = len(returns_sample2['AAPL']) - 1
n_spy_2 = len(returns_sample2['SPY']) - 1

print("AAPL's degrees of freedom: ", n_aapl_1)
print("SPY's degrees of freedom: ", n_spy_2)
```

```python
from scipy.stats import f

upper_critical_value = f.ppf(0.975, n_aapl_1, n_spy_2)
lower_critical_value = f.ppf(0.025, n_aapl_1, n_spy_2)

print("Upper critical value( given a = 0.05, n_1 = {0}, and n_2 = {1}): ".format(n_aapl_1, n_spy_2), upper_critical_value)
print("Lower critical value( given a = 0.05, n_1 = {0}, and n_2 = {1}): ".format(n_aapl_1, n_spy_2), lower_critical_value)
```

The $F$-statistic value is greater than the upper critical value for the $F$ test. Thus, the null hypothesis is rejected in favor of the alternative and conclude that the variances of AAPL and SPY are different.
