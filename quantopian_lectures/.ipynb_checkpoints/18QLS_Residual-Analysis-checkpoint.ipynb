# 18. QLS - Residual Analysis

## Linear Regression

Linear regression is used to estimate a linear relationship between a set of independent variables $X_{i}$ and a dependent outcome variable $y$. Our model takes the form of:

$$y_{i} = \beta_{0}1 + \beta_{i,1}x_{i,1} + \dots + \beta_{i,p}x_{i,p} + \epsilon_{i} = x_{i}'\beta + \epsilon_{i}$$

For $i \in \{1, \dots, n\}$, where $n$ is the number of observations. In vector form, this is:

$$y = X \beta + \epsilon$$

Where $y$ is a $n \times 1$ vector, $X$ is a $n \times p$ matrix, $\beta$ is a $p \times 1$ vector of coefficients, and $\epsilon$ is a standard normal error term. A model with $p = 1$ is a simple linear regression, while $p > 1$ is a multiple linear regression model.

The residuals of a model are the gaps between what the model predicts, and what is observed in sample. They can be used to check for some basic assumptions, such as:

- **Linear Fit**: The underlying relationship should be linear.
- **Homoscedastic**: The data should have no trend in the variance.
- **Independent and Identically Distributed**: The residuals of the regression should be independent and identically distributed (i.i.d.) and show no signs of serial correlation.

The residuals can diagnose whether the relationship estimated is real or spurious.

Statistical error is a similar metric, with one main difference: residuals quantify the gap between the model's predictions and the observed sample, while a statistical error is the difference between the model and the unobservable expected value. So, residuals are used to estimate this error.

```python
import numpy as np
import seaborn as sb
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.set_style("whitegrid", {"grid.color":".6", "grid.linestyle":":"})
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.stats.diagnostic as smd
import yfinance as yf
```

### Simple Linear Regression

First we define a function that performs linear regression and plots the results.

```python
def linreg(x, y):
    # running the linear regression
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    b0 = model.params[0]
    b1 = model.params[1]
    x = x[:, 1]

    # return summary of the regression and plot results
    sb.regplot(x=x, y=y).set(xlabel="X Values", ylabel="Y Values")
    return model, b0, b1
```

Now, define a toy relationship between $X$ and $Y$ to model with linear regression.

```python
n = 50
X = np.random.randint(0, 100, n)
epsilon = np.random.normal(0, 1, n)

Y = 10 + 0.5 * X + epsilon

linreg(X,Y)[0];
```

This toy example has some noise, as all real data will have. As a result, the 'line of best fit' will never *exactly* fit the data. Having a model that does fit every observation is most likely **overfit**.

For all models, there'll be a difference between the regression model's prediction, and what is observed in sample. These are residuals.

## Residuals

A residual is the difference between the regression model's predictions, and what is observed in sample. For any residual $r_{i}$, we express this as:

$$r_{i} = Y_{i} - \hat{Y_{i}}$$

Where $Y_{i}$ is the observed $Y$ value and $\hat{Y_i}$ is the *predicted* $Y$ value. The residuals can be pulled *out* of the model for analysis.

```python
model, B0, B1 = linreg(X,Y)
residuals = model.resid
print(residuals)
```

### Diagnosing Residuals

Most assumptions needed for a **valid** linear regression model can be checked through patterns in residuals. With a residual plot, the predicted values of the model and the residuals are compared.

```python
sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals");
```

A random distribution of residuals is ideal, meaning a linear model is likely a good fit. The points should form no pattern, if they do this might indicate autocorrelation or heteroscedasticity in the model.

### Appropriateness of a Linear Model

A plain linear regression model wouldn't be appropriate for an underlying relationship of the form:

$$Y = \beta_{0} + \beta_{1} X^{2}$$

... as a linear function wouldn't be able to explain the relationship between $X$ and $Y$.

In general, the residual plots of a linear regression on a non-linear relationship will show bias and be **asymmetrical** with respect to residual = $0$ line, linear relationship **will** be asymmetrical over the residual = $0$ axis.

E.g. consider a new relationship between $X$ and $Y$ that incorporates a quadratic term.

```python
n = 50
X = np.random.randint(0, 50, n)
epsilon = np.random.normal(0, 1, n)
Y_nonlinear = 10 - X**1.2 + epsilon

model = sm.OLS(Y_nonlinear, sm.add_constant(X)).fit()
B0, B1 = model.params
residuals = model.resid

print("β₀:", B0)
print("β₁:", B1)

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals");
```

The "inverted-U" shape is a sign that a non-linear model is needed.

## Heteroscedasticity

One of the main assumptions behind a linear regression is that the data is homoscedastic (i.e. the variance is constant). If some parts of the data have a different variance to other parts, then linear regression isn't appropriate. **Heteroscedasticity** refers to data with non-constant variance, as opposed to homoscedasticity.

Significant homoscedasticity invalidates linear regression by biasing the standard error of the model, meaning the outcomes of significance tests and confidence intervals can't be trusted. To avoid this, it's smart to use residual plots to check for heteroscedasticity and adjust.

E.g. consider another relationship between $X$ and $Y$.

```python
n = 50
X = np.random.randint(0, 100, n)
epsilon = np.random.normal(0, 1, n)
Y_heteroscedastic = 100 + 2*X + epsilon*X

model = sm.OLS(Y_heteroscedastic, sm.add_constant(X)).fit()
B0, B1 = model.params
residuals = model.resid

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals");
```

Heteroscedasticity often manifests as this spread, showing a *tapered* cloud in a direction. As it moves along the $x$-axis, the magnitude of the residuals increases. A linear regression can't cope with varying variance, hence the standard errors will be biased.

### Statistical Methods for Detecting Heteroscedasticity

Generally, it's good practice to back up qualitative observations on a plot with a quantitative method. The residual plots leads us to believe that the data is heteroscedastic. Let's confirm that with a statistical test.

A common way to test is using the *Breusch-Pagan* hypothesis test. A function exists in the `statsmodels` package called `het_breuschpagan` that simplifies the computation:

```python
breusch_pagan_p = smd.het_breuschpagan(model.resid, model.model.exog)[1]
print(breusch_pagan_p)
if breusch_pagan_p > 0.05:
    print("The relationship is not heteroscedastic.")
if breusch_pagan_p < 0.05:
    print("The relationship is heteroscedastic.")
```

The confidence level is set at $\alpha = 0.05$, so a Breusch-Pagan p-value below $0.05$ tells us the relationship is heteroscedastic. Multiple tests can be used to confirm that the results aren't false positives/negatives.

#### Differences Analysis

It involves looking at the first-order differences between adjacent values. Meaning, the changes from period to period of an independent variable are looked at, rather than looking directly at its values. By looking at the differences instead of the raw values, heteroscedasticity might be removed. Meaning the model can be corrected and used on the differences.

```python
# finding first-order differences in Y_heteroscedastic
Y_heteroscedastic_diff = np.diff(Y_heteroscedastic)
```

Now that the first-order differences are stored, the regression is repeated to check if there is still heteroscedasticity.

```python
model = sm.OLS(Y_heteroscedastic_diff, sm.add_constant(X[1:])).fit()
B0, B1 = model.params
residuals = model.resid

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals");
```

```python
breusch_pagan_p = smd.het_breuschpagan(residuals, model.model.exog)[1]
print(breusch_pagan_p)
if breusch_pagan_p > 0.05:
    print("The relationship is not heteroscedastic.")
if breusch_pagan_p < 0.05:
    print("The relationship is heteroscedastic.")
```

*Note: The new regression was calculated on the **differences** between data, therefore the regression output must be back-transformed to reach a prediction in the original scale. Since the differences were regressed, they can be added to the original data to get an estimate:*

$$\hat{Y_{i}} = Y_{i-1} + \hat{Y_{d}}$$

#### Logarithmic Transformation

A log transformation brings residuals closer and, ideally, removes heteroscedasticity. In many (though not all) cases, it's sufficient in stabilizing the variance of a relationship.

```python
# taking the log of the previous data Y_heteroscedastic and saving it in Y_heteroscedastic_log
Y_heteroscedastic_log = np.log(Y_heteroscedastic)
```

Now that the log-transformed versions are stored, the regression is repeated to check if there is still heteroscedasticity.

```python
model = sm.OLS(Y_heteroscedastic_log, sm.add_constant(X)).fit()
B0, B1 = model.params
residuals = model.resid

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals")

breusch_pagan_p = smd.het_breuschpagan(residuals, model.model.exog)[1]
print(breusch_pagan_p)
if breusch_pagan_p > 0.05:
    print("The relationship is not heteroscedastic.")
if breusch_pagan_p < 0.05:
    print("The relationship is heteroscedastic.")
```

*Note: The new regression was calculated on the **log** of the original data, therefore the regression output must be back-transformed to reach a prediction in the original scale. Since the scale has been altered, it can be transformed to the original data using the inverse of the log:*

$$\hat{Y} = e^{\log(\hat{Y})}$$

#### Box-Cox Transformation

The Box-Cox transformation tests all values of $\lambda$ within the range $[-5,5]$ to see which makes the output of the following equation closest to being normally distributed:

$$Y^{(\lambda)} = \begin{cases}
  \frac{Y^{\lambda}-1}{\lambda} & : \lambda \neq 0\\ \log{Y} & : \lambda = 0
\end{cases}$$

The "best" $\lambda$ will be used to transform the series along the function.
*Note: the Box-Cox transformation can **only** be used if all the data is positive.*

```python
# finding a power transformation adjusted Y_heteroscedastic
Y_heteroscedastic_box_cox = stats.boxcox(Y_heteroscedastic)[0]
```

Now that the power-transformed versions are stored, the regression is repeated to check if there is still heteroscedasticity.

```python
model = sm.OLS(Y_heteroscedastic_box_cox, sm.add_constant(X)).fit()
B0, B1 = model.params
residuals = model.resid

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals")

breusch_pagan_p = smd.het_breuschpagan(residuals, model.model.exog)[1]
print(breusch_pagan_p)
if breusch_pagan_p > 0.05:
    print("The relationship is not heteroscedastic.")
if breusch_pagan_p < 0.05:
    print("The relationship is heteroscedastic.")
```

*Note: The new regression was calculated to $x$ power of the original data, therefore the regression output must be back-transformed to reach a prediction in the original scale. Since the scale has been altered, it can be transformed to the original data using the inverse of the Box-Cox function:*

$$\hat{Y} = \left( Y^{(\lambda)} \lambda + 1 \right)^{ \frac{1}{\lambda}}$$

## Autocorrelation

Another assumption behind linear regression is that the residuals aren't autocorrelated. A series is said to be autocorrelated when it is correlated with a **delayed** version of itself. E.g. daily high temperatures are autocorrelated, as today's temperature gives a reasonable estimate for tomorrow's temperature. On the other hand, a series of die rolls aren't autocorrelated as seeing one roll does not give information on what the next might be. Each roll is independent.

Stock prices are usually autocorrelated, while returns are independent. A time dependency on previous values is represented like so:

$$Y_{i} = Y_{i-1} + \epsilon$$

If the residuals of a model are autocorrelated it means predictions can be made about adjacent -or future- residuals.

```python
n = 50
X = np.linspace(0, n, n)
Y_autocorrelated = np.zeros(n)
Y_autocorrelated[0] = 50
for t in range(1, n):
    Y_autocorrelated[t] = Y_autocorrelated[t-1] + np.random.normal(0, 1)

# regressing X and Y_autocorrelated
model = sm.OLS(Y_autocorrelated, sm.add_constant(X)).fit()
B0, B1 = model.params
residuals = model.resid

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals");
```

### Statistical Methods for Detecting Autocorrelation

As with all statistical properties, a test is required to decide whether autocorrelation is present. To this end, a *Ljung-Box* test is good.

A Ljung-Box test examines autocorrelation at all lag intervals below a specified maximum, and return, arrays containing the outputs for every tested lag interval.

The tests can be carried out using the `acorr_ljungbox` function in `statsmodels`.

```python
ljung_box = smd.acorr_ljungbox(residuals, lags=[10], return_df=False)
print("Lagrange Multiplier Statistics:", ljung_box["lb_stat"].tolist())
print("P-values:", ljung_box["lb_pvalue"].tolist(), "\n")

if any(ljung_box["lb_pvalue"] < 0.05):
    print("The residuals are autocorrelated.")
else:
    print("The residuals are not autocorrelated.")
```

Since the test yielded a p-value below 0.05 for **at least** one lag interval, the residuals are autocorrelated.

### Adjusting for Autocorrelation

Adjusting for autocorrelation is similar to adjusting for heteroscedasticity. First, check if a model on the first-order differences of $Y$ has autocorrelated residuals.

```python
# finding first-order differences in Y_autocorrelated
Y_autocorrelated_diff = np.diff(Y_autocorrelated)

model = sm.OLS(Y_autocorrelated_diff, sm.add_constant(X[1:])).fit()
B0, B1 = model.params
residuals = model.resid

sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals")

ljung_box = smd.acorr_ljungbox(residuals, lags=[10], return_df=False)
print("Lagrange Multiplier Statistics:", ljung_box["lb_stat"].tolist())
print("P-values:", ljung_box["lb_pvalue"].tolist(), "\n")

if any(ljung_box["lb_pvalue"] < 0.05):
    print("The residuals are autocorrelated.")
else:
    print("The residuals are not autocorrelated.")
```

*Note: The new regression was calculated on the **differences** between data, therefore the regression output must be back-transformed to reach a prediction in the original scale. Since the differences were regressed, they can be added to the original data to get an estimate:*

$$\hat{Y_{i}} = Y_{i-1} + \hat{Y_{d}}$$

Log transformations can be performed the same way as they are performed for heteroscedasticity.

## Example: Market Beta Calculation

Calculate the market beta between TSLA and SPY using simple linear regression, then conduct a residual analysis on the regression to ensure validity. SPY returns will be the independent variable, while TSLA returns are the outcome variable. The regression will yield a line of best fit:

$$\hat{r_{TSLA}} = \hat{\beta_{0}} + \hat{\beta_{1}}r_{SPY}$$

The slope of the regression line, $\hat{\beta_{1}}$, will represent market beta: for every $r$ percent change in SPY the predicted returns of TSLA will change by $\hat{\beta_{1}}$.

```python
start = "2014-01-01"
end = "2015-01-01"

benchmark = yf.download("SPY", start, end)
benchmark = benchmark["Adj Close"]
asset = yf.download("TSLA", start, end)
asset = asset["Adj Close"]

# get rid of the first (0th) element because it is NAN to get returns
r_a = asset.pct_change()[1:].values
r_b = benchmark.pct_change()[1:].values

# regressing the benchmark b and asset a
r_b = sm.add_constant(r_b)
model = sm.OLS(r_a, r_b).fit()
r_b = r_b[:, 1]
B0, B1 = model.params

# plotting the regression
sb.regplot(x=r_b, y=r_a).set(xlabel="TSLA Returns", ylabel="SPY Returns")

# print our result
print("Estimated TSLA β:", B1)

# calculating the residuals
residuals = model.resid
```

The regression yielded an estimated beta of 1.9541; according to the regression, for every 1% return in SPY, there should be an 1.95% return in TSLA.

Now that the regression results and residuals are done, analysis must be conducted.

```python
sb.residplot(x=model.predict(), y=residuals, lowess=True, color="orange").set(xlabel="Predicted values", ylabel="Residuals");
```

By just looking at the distribution of the residuals it's clear that abnormalities aren't present. The distribution is relatively random and there are no patterns (the clustering around the origin is a result of returns clustering around 0). The qualitative conclusion is that residuals are homoscedastic and not autocorrelated, therefore satisfying the assumptions for linear regression.

### Breusch-Pagan Heteroscedasticity Test

```python
bp_test = smd.het_breuschpagan(model.resid, model.model.exog)

print("Lagrange Multiplier Statistic:", bp_test[0])
print("P-value:", bp_test[1])
print("f-value:", bp_test[2])
print("f_p-value:", bp_test[3], "\n")
if bp_test[1] > 0.05:
    print("The relationship is not heteroscedastic.")
if bp_test[1] < 0.05:
    print("The relationship is heteroscedastic.")
```

Since the p-value is greater than 0.05 it means that there isn't evidence to reject the null hypothesis, therefore the relationship is homoscedastic.

### Ljung-Box Autocorrelation Test

```python
ljung_box = smd.acorr_ljungbox(r_a, lags=[10], return_df=False)
print("Lagrange Multiplier Statistics:", ljung_box["lb_stat"].tolist())
print("P-values:", ljung_box["lb_pvalue"].tolist(), "\n")

if any(ljung_box["lb_pvalue"] < 0.05):
    print("The residuals are autocorrelated.")
else:
    print("The residuals are not autocorrelated.")
```

Since the p-value is greater than 0.05 it means that the residuals aren't autocorrelated.
